```python
import chromadb
import pandas as pd
from sentence_transformers import SentenceTransformer
import torch

# Configuración
CHROMA_DB_PATH = "./data/processed/chroma_db"
CSV_PATH = "./data/processed/climateTwitterData_sentiments_updated.csv"
MODEL_NAME = "sentence-transformers/all-MiniLM-L6-v2"
```

```python
# Inicializar el cliente de ChromaDB
client = chromadb.PersistentClient(path=CHROMA_DB_PATH)
collection = client.get_collection(name="climate_tweets")

# Cargar el modelo de embedding
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = SentenceTransformer(MODEL_NAME).to(device)

# Cargar los datos
df = pd.read_csv(CSV_PATH)
```

```python
def generate_embedding(text):
    """Genera un embedding para el texto dado."""
    return model.encode(text, convert_to_tensor=True).cpu().numpy().tolist()

def search_tweets(query, n_results=5):
    """Busca tweets relacionados con la consulta."""
    query_embedding = generate_embedding(query)
    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=n_results
    )
    return results

def display_results(results):
    """Muestra los resultados de la búsqueda."""
    num_results = 0
    for i, (id, text, distance) in enumerate(zip(results['ids'][0], results['documents'][0], results['distances'][0]), 1):
        print(f"\nResultado {i}:")
        print(f"ID: {id}")
        print(f"Texto: {text}")
        print(f"Distancia: {distance}")
        tweet_data = df[df.index.astype(str) == id].iloc[0]
        print(f"Sentimiento: {tweet_data['sentiment1']}")
        num_results+=1
    print(f"Numero de resultados de la busqueda: {num_results}")

def interactive_search():
    """Función para realizar búsquedas interactivas."""
    print("Sistema de Recuperación de Información de Tweets sobre Clima")
    print("Escriba 'salir' para terminar.")
    while True:
        query = input("\nIngrese su consulta: ")
        if query.lower() == 'salir':
            break
        n_results = int(input("\nIngrese cuentos resultados quiere visualizar: "))
        print(query)
        results = search_tweets(query, n_results)
        display_results(results)

if __name__ == "__main__":
    interactive_search()
```
